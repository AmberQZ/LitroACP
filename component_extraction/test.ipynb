```python
import subprocess

process = subprocess.Popen(['python', './gilacp.py', '--mode', 'acre','--acr','True'], stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)

for line in process.stdout:
    print(line, end='')  # 保持流式输出
process.wait()
```
